### spaCy Tokenization Process:
- Duiring preprocessing, spacy first tokenizes the text, i.e. it segments it into words, punctuation and so on. This is done by applying rules specific to each language.
- For example, punctuation at the end of a sentence should be split off. 
- First, the raw text is split on whitespace characters to text.split(' ').
- Then, the tokenizer processes the text from left to right to perform two checks.

```python
### General model

def helper(nlp, text):
    doc = nlp(text)

    ### chop down the text into tokens 
    ### spaCy's library allows us to tokenize our text via sentence tokenization & word tokenization
    lst = []
    for token in doc:
        if token.is_stop == False and token.text.isalpha() == True: # remove stopwords & punctuations
            if token.lemma_ not in lst:
                lst.append(token.lemma_)
    return lst


def multi_lingual_text_preprocessing(language, language_package):
    nlp = spacy.load(language_package)

    # DEV Short Description
    short_description_language = data[data['language_short_description']==language]
    short_description_language['preprocessed_short_description'] = short_description_language['DEV Short Description'].apply(lambda x: helper(nlp, x))

    # INV Executive Summary
    executive_summary_language = data[data['language_executive_summary']==language]
    executive_summary_language['preprocessed_executive_summary'] = executive_summary_language['INV Executive Summary (En)'].apply(lambda x: helper(nlp, x))

    return short_description_language, executive_summary_language
```

### Clustering Methods

#### Kmeans
Model Parameter:
- Run K-Means for many different "number of clusters" values and score ach clustering with some "cluster goodness" measure (usually a variation on intra-cluster vs inter-cluster distances) and attempt to find an "elbow"

Pros:
- Fast, easy to understand and use
- With right tricks and optimizations, can be made exceptionally efficient and have excellent performance
- For huge data, KMeans might be the only option

Cons: 
- Require to first select # of clusters 
- Suffer if the data is not spherical
- Be vulnerable to outliers; depends on initialization; may not be stable 


#### Agglomerative Clustering
- Start with each point in its own cluster. For each cluster, use some criterion to choose another cluster to merge with; do this repeatedy until have only one cluster and get a hierarchy or binary tree of clusters branching down to the last layer.

#### Density-based Spatial Clustering of Applications with Noise/ DBSCAN 
Model:
- Density-based algorithm: it assumes clusters for dense regions
- The first actual clustering algorithm; don't require every point to be assigned to a cluster and hence don't partition the data; but instead extracts the "dense" clusters and leave sparse background classified as "noise"
- Intuitive parameters: epsilon as a distance value 
- Can tackle large datasets

Pros:
- Avoid outliers; stable across runs
- Work well with arbitrarily shaped and sized clusters 
- Don't have centroids, the clusters are formed by linking neighbor points together 

Cons:
- Epsilon can be hard to pick. The algorithm can be quite sensitive to the choice of the parameter
- The density based transformation depends on another parameter (min_samples/choice of clustering)


#### Hierarchical Density-based Spatial Clustering of Applications with Noise/ HDBSCAN
Model:
- Recent algorithm improved based on DBSCAN, to allow varing density clusters
- Starts of similarily as DBSCAN: transform the space according to density; perform single linkage clustering on the transformed space
- Instead of taking an epsilon value as a cut level for the dendrogram; the dendropgram is condensed by viewing splits that result in a small number of points splitting off as points "falling out of a cluster". min_cluster_size is used to determine whether points are "falling out of a cluster" or splitting to form two new clusters 

Cons:
- min_samples is not intuitive. HDBSCAN is not that sensitive to this and we can choose sensible defaults. But this remains the biggest weakness of the algorithm

#### Steps of HDBSCAN:
1. Transform the space according to density/sparsity
- Assumptions of noise is important. A single noise data point in the wrong place can act as a bridge between islands, gluing them together. 
- To characterize "sea" and "land" without doing a clustering, we get an estimate of density where we consider lower density points as the "sea". For practical use, we make "sea" points more distant from each other and from the "land".
- An inexpensive estimate of density: mutual reachability distance. d_mrk(a,b)=max{core_k(a), core_k(b), d(a,b)}


2. Build the minimum spanning tree of the distance weighted graph
- Now that we have a new mutual reachability distance on the data, we start finding the islands on dense data. Dense areas are relative; different islands may have different densities. 
- Consider the data as a weighted graph with the data points as vertices and an edge between any two points with weight equal to the mutual reachability distance of those points.
- Now consider a threshold value, starting high and steadily being lowered. Drop any edges with weight above that threshold. Eventually we will have a hierarchy of connected components at varing threshold levels.
- Graph theory furnishes us with such a thing: the minimum spanning tree of the graph. We build the tree one edge at a time, always adding the lowest weight edge that connects the current tree to a vertex not yet in the tree. 


3. Construct a cluster hierarchy of connected components
- Given the minimum spanning tree, the next step is to convert that into the hierarchy of connected components. Sort the edges of the tree by distance in increasing order; iterate through and creat a new merged cluster for each edge. 


4. Condense the cluster hierarchy based on minimum cluster size
- The first step in cluster extraction is condensing down the large and complicated cluster hierarchy into a smaller tree with a little more data attached to each node.
- Rather than seeing it as a cluster splitting into two new clusters, we declare it to be "points falling out of a cluster" 


5. Extract the stable clusters from the condensed tree


#### Things to Mention:
- HDBSCAN is noise aware – it has a notion of data samples that are not assigned to any cluster. This is handled by assigning these samples the label -1.

- The hdbscan library implements soft clustering, where each data point is assigned a cluster membership score ranging from 0.0 to 1.0. A score of 0.0 represents a sample that is not in the cluster at all (all noise points will get this score) while a score of 1.0 represents a sample that is at the heart of the cluster (note that this is not the spatial centroid notion of core). You can access these scores via the probabilities_ attribute.

### Topic Modeling

- A type of statistical modeling tool; unsupervised maching learning method; 


##### LSA(Latent Semantic Analysis)
- Use bag-of-word and term-document matrix to detect topic
- Assumption: different topics have different distributions of words in them

##### Steps of LSA
- Create word-document frequency matrix, where in the rows we have words, in the columns we have the different documents. A cell (i,j) of the matrix denotes how many times the word w_i has occured in the document D_j.
- Factorize the matrix into three matrices
1. word-topic frequency matrix: in this matrix, words are in row, topics are in columns. Each cell denotes the occurence of a word in the respective column. [M*N]
2. topic importance matrix: this is a diagnal matrix, where the ith diagnal element of the matrix denotes the importance of the topic i. [N*N]
3. topic document matrix: Topics are in rows, documents are in columns, each cell denotes the weight of a topic in the respective document. [N*M]

Using SVD, we factorize matrix M as follows:
- M=UDV


##### LDA(Latent Dirichlet Analysis)
- Significant improvement from LSA (LSA considers no probabilistic determination inside the document structures)

- LDA is a generative probabilistic model of a corpus. The basic idea is that the document are represented as random mixtures over latent topics, where each topic is characterized by a distribution over words. 

- One application of LDA in machine learning, specifically, topic discovery, a subproblem in natural language processing, is to discover topics in a collection of documents, and then automatically classify any individual document whithin the collection in terms of how "relevant" it is to each of the discovered topics. A topic is considered to be a set of terms that taken together, suggest a shared theme. 

##### Steps of LDA
- Choose N (which follows a possion distribution with lambda=e)
- Choose theta (which follows a dirichlet distribution with alpha=a)
- For each word w_n: choose a topic z_n, which follow multinomial distribution with theta, choose a word w_n which comes from p(w_n|z_n, beta), which is a multinomial probability conditioned on topic z_n
- To finally solve the problem, optimize alpha and beta for decreasing the kullback leibler divergence distance between the theoretical and observed probability distributions
